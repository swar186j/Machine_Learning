```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import pandas as pd
```

```python
from sklearn import datasets
from sklearn.decomposition import PCA
```

```python
#load the data set
iris=datasets.load_iris()
```

```python
x= iris.data    #input variable
y= iris.target  #target value
```

```python
x.shape     #size of the dataset
```

```python
y.shape
```

```python
#used for reducing the dimensionality of x from 4 to 2 components
pca = PCA(n_components=2)
```

```python
#for creating principle component of the data we use fit()
pca.fit(x)
```

```python
#prints the eigen vectors of covariance matrix of x set of observations
pca.components_
```

```python
#transform x to feature component space
z= pca.transform(x) 
```

```python
#hence pca reduced number of dimensions from 4 to 2
z.shape
```

```python
#plot classes with diff colors using c parameter
plt.scatter(z[:,0], z[:,1], c=y) 
```

```python
#cumulative explained variance ratio
pca.explained_variance_ratio_    
```
